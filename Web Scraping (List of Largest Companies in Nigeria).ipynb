# List of Largest Companies in Nigeria
## By Revenue

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_Nigeria"

page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, "html")
print(soup)
```

```python
soup.find("table")
```

```python
table = soup.find("table")
print(table)
```

```python
table.find_all("th")
```

```python
company_titles = table.find_all("th")
company_titles
```

```python
company_table_titles = [title.text.strip() for title in company_titles]

print(company_table_titles)
```

```python
import pandas as pd

df = pd.DataFrame(columns=company_table_titles)
df
```

```python
column_data = table.find_all("tr")

for row in column_data[1:]:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
    
    
```
